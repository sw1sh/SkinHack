```python
import sugartensor as tf
import numpy as np
import pandas as pd
```

```python
sess = tf.InteractiveSession()
```

```python
train_size = 8522
batch_size = 2   # batch size

csv_reader = tf.TextLineReader(skip_header_lines=1)
csv_file = tf.train.string_input_producer(['/volume/data/train_age_full_2.txt'])
_, csv_row = csv_reader.read(csv_file)

record_defaults = record_defaults = [
                   tf.constant([], dtype=tf.string),    # Column 0
                   tf.constant([20], dtype=tf.int32),     # Column 1
                   tf.constant(['m'], dtype=tf.string),    # Column 2
                   tf.constant([], dtype=tf.string)]    # Column 3
_, age, gender, image_path = tf.decode_csv(csv_row, record_defaults=record_defaults)
                            
image_queue = tf.FIFOQueue(10000, tf.string)
age_queue = tf.FIFOQueue(10000, tf.int32)
image_init = image_queue.enqueue(image_path)
age_init = age_queue.enqueue(age)

#filename_queue = tf.train.string_input_producer(image_file)
image_reader = tf.WholeFileReader()

_, image_data = image_reader.read(image_queue)
image = tf.image.decode_jpeg(image_data)

image_slice = tf.slice(image, [0,0,0], [1505,2257,3])
image_slice.set_shape((1505, 2257, 3))

images = tf.train.batch([image_slice], batch_size=batch_size, num_threads=1, capacity=3*batch_size, dynamic_pad=True,allow_smaller_final_batch=True)
ages = tf.train.batch([age], batch_size=batch_size, num_threads=1, capacity=3*batch_size, dynamic_pad=False,allow_smaller_final_batch=True)
```

```python
tf.to_float(images).sg_conv(dim=64).sg_flatten().sg_dense(dim=128)
```

```python
y = tf.to_float(images).sg_conv(dim=64).sg_conv(dim=128).sg_flatten().sg_dense(dim=1024).sg_dense(dim=128).sg_dense(dim=1, act='linear')
```

```python
err = y.sg_mse(target=tf.to_float(ages))
```

```python
opt = tf.sg_optim(err, lr=0.0001)
```

```python
sess.run(tf.initialize_all_variables())
```

```python
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(coord=coord)

image_qr = tf.train.QueueRunner(image_queue, [image_init])
image_threads = image_qr.create_threads(sess, coord=coord, start=False)

age_qr = tf.train.QueueRunner(age_queue, [age_init])
age_threads = age_qr.create_threads(sess, coord=coord, start=False)
```

```python
for step in range(4):
    if coord.should_stop():
        break
        
    for i in range(batch_size):
        sess.run(image_init)
        sess.run(age_init)
        
    l = sess.run([err, opt])[0]
    print np.mean(l)
    #print np.shape(sess.run(images))

#coord.request_stop()
#coord.join(threads)
#coord.join(image_threads)
```

```python
#
# training
#

# def alternate training func
@tf.sg_train_func
def alt_train(sess, opt):
    l = sess.run([err, opt])[0]
    return np.mean(l)

# do training
alt_train(log_interval=10, max_ep=30, ep_size=train_size // batch_size, early_stop=False)
```
